```python
import pandas as pd
import numpy as np
import re
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

%matplotlib inline
plt.style.use('fivethirtyeight')

df= pd.read_csv('C:\\Users\\USER\\Desktop\\countries-aggregated_csv.csv', parse_dates= ['Date'])

df.head(4)
```

```python
df['Total_cases'] = df[['Confirmed','Deaths','Recovered']].sum(axis= 1)
df.head()
```

```python
#worldwide cases

worldwide_df = df.groupby(['Date']).sum()
w = worldwide_df.plot(figsize=(12,6))
w.set_xlabel('Date')
w.set_ylabel('NO. of Cases Worldwide')
w.title.set_text('Worldwide covid insights')

plt.show()
```

```python
#visualizing the data using pandas and matplotlib
Is_df = df[df['Country']== 'India'].groupby(['Date']).sum()
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)

ax.plot(worldwide_df[['Total_cases']],label ='worldwide')
ax.plot(Is_df[['Total_cases']], label ='India')

ax.set_xlabel('Date')
ax.set_ylabel('NO. of Total_Cases ')
ax.title.set_text('Worldwide vs India Total_cases')

plt.legend(loc ='upper left')
plt.show()
```

```python
#Israel Daily Deaths and Cases
Is_df = Is_df.reset_index()
Is_df.head()
```

```python
Is_df['Daily Confirmed']= Is_df['Confirmed'].sub(Is_df['Confirmed'].shift())
Is_df['Daily Deaths'] = Is_df['Deaths'].sub(Is_df['Deaths'].shift())
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)

ax.bar(Is_df['Date'],Is_df['Daily Confirmed'], color='b', label='India Daily Confirmed Cases')
ax.bar(Is_df['Date'],Is_df['Daily Deaths'],color ='r',label ='India Daily Deaths cases')
ax.set_xlabel('Date')
ax.set_ylabel('No. of people Affected')
ax.title.set_text('India Daily Cases and Deaths')

plt.legend(loc ='upper left')
plt.show()
```
